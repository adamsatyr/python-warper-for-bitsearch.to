```python
import requests
from bs4 import BeautifulSoup
import json
```

```python
query = 'linux iso'
query = query.replace(" ", "+")
```

```python
# list(soup.children)

torrent_list = []
for page_number in range(0, 1):

    url = 'https://bitsearch.to/search?q={}"&page={}&sort=seeders"'.format(query, page_number)

    page = requests.get(url)
    soup = BeautifulSoup(page.content, 'html.parser')
    search_box = soup.select("div.search-result.view-box")
    
    for item in range(0, len(search_box)):
        params_list = list(filter(None, search_box[item].find_all('div')[0].get_text().split('\n')))
        torrent_list.append({
            'Name':params_list[0],
            'Size': params_list[3],
            'Downloads': params_list[2],
            'Seeders': params_list[4],
            'Leechers': params_list[5],
            'DateUploaded': params_list[6],
            'Url': 'https://bitsearch.to{}'.format(search_box[item].select('.info h5 a')[0]['href']),
            'TorrentLink': search_box[item].select('.links a')[0]['href'],
            'Magnet':  search_box[item].select('.links a')[1]['href']
        })

    
    # for item in torrent_list:
    #     print(item)
```

```python
print(json.dumps(torrent_list, indent=4, sort_keys=False))
```
